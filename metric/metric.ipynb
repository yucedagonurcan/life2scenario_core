```python
def find_verb(prompt):
    verbs = ['add', 'remove']
    if verbs[0].lower() in prompt.lower():
        return verbs[0].lower()
    elif verbs[1].lower() in prompt.lower():
        return verbs[1].lower()
```

```python
def find_object(prompt, verb):
    words = prompt.split()
    target_index = words.index(verb)
    return words[target_index + 1]
```

```python
def count_object(code, target_word):
    pattern = r'\b' + re.escape(target_word) + r'\b'
    matches = re.findall(pattern, code, re.IGNORECASE)
    return len(matches)
```

```python
def is_count_match(prompt, ground_truth, prediction):
    verb = find_verb(prompt)
    gt_count = count_object(ground_truth, verb) + 1
    pred_count = prediction
    if gt_count == pred_count:
        return 1
    else:
        return 0
```
